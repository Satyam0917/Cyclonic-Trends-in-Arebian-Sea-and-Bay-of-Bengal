```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'your_data.csv' is the name of your dataset file
file_path = '/content/BOB_Cyclones_1982_2023.csv'

# Load the dataset into a pandas DataFrame
df = pd.read_csv(file_path)

# Convert the 'date' column to datetime format, letting pandas infer the format
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Filter entries where Time (UTC) is 1200
df_1200 = df[df['Time (UTC)'] == 1200]

# Sort the DataFrame based on Cyclone and Date in ascending order
sorted_df = df_1200.sort_values(by=['Cyclone', 'date'])

# Sort the DataFrame based on index in ascending order
sorted_df = sorted_df.sort_index()

# Display the sorted DataFrame
pd.set_option("display.max_rows", None, "display.max_columns", None)

# Plotting a horizontal bar graph
plt.figure(figsize=(12, 8))

# Assign different colors based on the index
colors = ['red' if i < 449 else 'blue' for i in sorted_df.index]

plt.barh(sorted_df['Cyclone'], sorted_df['Estimated Maximum Sustained Surface speed (kt)'], color=colors)

# Adding labels and title
plt.xlabel('Intensity (kt)')
plt.ylabel('Cyclones and Severe Cyclones in Bay of Bengal')

# Add legends
red_patch = plt.Line2D([0], [0], marker='o', color='w', label='Cyclones before 2001', markerfacecolor='red', markersize=10)
blue_patch = plt.Line2D([0], [0], marker='o', color='w', label='Cyclones after 2001', markerfacecolor='blue', markersize=10)

plt.legend(handles=[red_patch, blue_patch])

plt.show()
```
